```python
import pandas as pd
```

```python
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))   #Here we fill the null values by making median for catagory wise because the rating can be diffrent from other categories like Cloths and Footwears
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns= {'purchase_amount_(usd)':'purchase_amount'})   #we do all the words in small letters and spaces into '_'
```

```python
df.columns
```

```python
#createing the new Columns age_group
labels=['young_adults','Adults','Middle_aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) 
```

```python
df[['age','age_group']].head(10)
```

```python
#create the column purchase_frequency_days

frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months': 90
}

df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df= df.drop('promo_code_used',axis=1)
```
